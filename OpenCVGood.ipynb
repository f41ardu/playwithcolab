Load an imag file and show image 


```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
import sys

img = cv.imread("opencv/butterfly.jpg")
```

```python
if img is None: 
  sys.exit((" Could not read image."))
else: 
  print("read image")
```

```python
imgGray= cv.cvtColor(img, cv.COLOR_RGB2GRAY)
hist = cv.calcHist([imgGray],[0],None,[256],[0,256])
eq = cv.equalizeHist(imgGray)
cdf = hist.cumsum()
cdfnmhist = cdf * hist.max()/ cdf.max()
histeq = cv.calcHist([eq],[0],None,[256],[0,256])
cdfeq = histeq.cumsum()
cdfnmhisteq = cdfeq * histeq.max()/ cdf.max()
cv2_imshow(img)
plt.subplot(221), plt.imshow(imgGray,'gray')
plt.subplot(222), plt.plot(hist), plt.plot(cdfnmhist)
plt.subplot(223), plt.imshow(eq,'gray')
plt.subplot(224), plt.plot(histeq), plt.plot(cdfnmhisteq)
plt.xlim([0,512])
```
